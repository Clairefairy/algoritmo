```python
for elefantes in [1, 3, 5, 7, 9]:
    if elefantes == 1 :
        print("1 elefante incomoda muita gente\n 2 elefantes incomodam, incomodam muito mais")
    else :
        print(elefantes, 'elefantes incomodam muita gente\n', elefantes+1,'elefantes' )
        print((elefantes+1)*'incomodam, ','muito mais')
print("\n")
for elefante in [10, 8, 6, 4, 2]:
    if elefante == 2 :
        print("2 elefantes incomodam pouca gente\n 1 elefante incomoda muito menos")
    else :
        print(elefante, 'elefantes incomodam pouca gente\n', elefante-1,'elefantes' )
        print((elefante-1)*'incomodam, ','muito menos')
```
